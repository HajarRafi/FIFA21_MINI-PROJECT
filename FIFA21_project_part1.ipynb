## Data Cleaning

```python
#standardize header names
data.columns = [i.lower().replace(" ", "_") for i in data.columns]
data.columns.values
```

```python
#remove contract from team, as there is a separate column for contract
data.rename(columns = {'team_&_contract' : 'team'}, inplace = True)
#define function
def remove_contract(x):
    for i in range(len(data)):  #to remove dates from team column
        if data['contract'][i] in x[i]:
            x[i] = x[i].replace(data['contract'][i], '')
        else:
            x[i] = x[i]
            
    for i in range(len(data)):  #and then, remove remained spaces at the end
        if x[i].endswith(' '):
            x[i] = x[i].rstrip(x[i][-1])
        else:
            x[i] = x[i]
    return x
```

```python
#height, weight, joined, value, wage, release_clause columns should be transformed
#they need to be numeric for future analysis
```

```python
#change inches to cm
def transform_height(x):
    h_ft = int(str(x).replace('"', '').split("'")[0])  #to separate feet and inches and then apply the formula
    h_inch = int(str(x).replace('"', '').split("'")[1])
    h_inch += h_ft * 12
    h_cm = round(h_inch * 2.54, 0)
    return int(h_cm)
#apply the function
data['height'] = data['height'].apply(transform_height)
```

```python
#change pounds to kg
def transform_weight(x):
    w_lbs = int(str(x).replace('lbs', ''))
    w_kg = round(w_lbs / 2.2046, 0)
    return int(w_kg)
#apply the function
data['weight'] = data['weight'].apply(transform_weight)
```

```python
#leave only the year of joining
def transform_joined(x):
    j_year = int(str(x).split(',')[1])
    return j_year
#apply the function
data['joined'] = data['joined'].apply(transform_joined)
```

```python
#define a function that will remove € sign and change K and M to 0's
def transform_money(x):
    x = str(x).replace('€', '')
    if 'M' in str(x):
        x = str(x).replace('M', '')
        x = float(x) * 1000000
    elif 'K' in str(x):
        x = str(x).replace('K', '')
        x = float(x) * 1000
    return int(x)
#apply it to value, wage and release_clause columns
data['value'] = data['value'].apply(transform_money)
data['wage'] = data['wage'].apply(transform_money)
data['release_clause'] = data['release_clause'].apply(transform_money)
```

```python
#it will be good to leave in contract column only the end year, as the beginning is already in joined column
data['contract'] = data['contract'].apply(lambda x: str(x)[-4:])
data.rename(columns = {'contract' : 'end_contract'}, inplace = True)  #and rename accordingly
```
